```python
dict_index = state_total_merge.to_dict(orient='index')

print(dict_index)
```

```python
data=dict_index


# Calculate revenue per capita for each state
for state in data.values():
    state['revenue_per_capita'] = state['revenue'] / state['population total 18+']

# Separate data into states with taxes and those without taxes
states_with_taxes = {key: values for key, values in data.items() if values['state_taxes_on_gambling'] == 'Yes'}
states_without_taxes = {key: values for key, values in data.items() if values['state_taxes_on_gambling'] == 'No'}

# Get top states by revenue per capita for both categories
top_states_with_taxes = sorted(states_with_taxes.values(), key=lambda x: x['revenue_per_capita'], reverse=True)[:5]
top_states_without_taxes = sorted(states_without_taxes.values(), key=lambda x: x['revenue_per_capita'], reverse=True)[:5]

plt.figure(figsize=(12, 8))

# Plot states with taxes
plt.subplot(1, 2, 1)
plt.bar([state['state'] for state in top_states_with_taxes], [state['revenue_per_capita'] for state in top_states_with_taxes], color='skyblue')
plt.xlabel('State')
plt.ylabel('Revenue Per Capita ($)')
plt.title('Top 5 States with Taxes by Revenue Per Capita')

# Plot states without taxes
plt.subplot(1, 2, 2)
plt.bar([state['state'] for state in top_states_without_taxes], [state['revenue_per_capita'] for state in top_states_without_taxes], color='lightcoral')
plt.xlabel('State')
plt.ylabel('Revenue Per Capita ($)')
plt.title('Top 5 States without Taxes by Revenue Per Capita')

plt.tight_layout()
plt.show()
```

```python
# Function to convert a CSV to JSON
# Takes the file paths as arguments
def make_json(csvFilePath, jsonFilePath):
     
    # create a dictionary
    data = {}
     
    # Open a csv reader called DictReader
    with open(csvFilePath, encoding='utf-8') as csvf:
        csvReader = csv.DictReader(csvf)
         
        # Convert each row into a dictionary
        # and add it to data
        for rows in csvReader:
             
            #primary key
            key = rows['']
            data[key] = rows
 
    # Open a json writer, and use the json.dumps()
    # function to dump data
    with open(jsonFilePath, 'w', encoding='utf-8') as jsonf:
        jsonf.write(json.dumps(data, indent=4))
```
