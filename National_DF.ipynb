```python
# Dependencies
import requests
import csv
import json
import pandas as pd
from pathlib import Path
```

```python
#read the csvs file into a data frame
national_data=Path('Resources/month_rev_each_state_df_FINALSET.csv')
national_df=pd.read_csv(national_data)
#national_df
```

```python
#Change the date column to datetime format
national_df=national_df.astype({'month':'datetime64[D]'})
#national_df
```

```python
#Change handle, revenue and taxes to dollar format
national_df['handle']=national_df['handle'].astype(float).map('${:,.2f}'.format)
national_df['revenue']=national_df['revenue'].astype(float).map('${:,.2f}'.format)
national_df['taxes']=national_df['taxes'].astype(float).map('${:,.2f}'.format)

#national_df
```

```python
#Change the hold column to show as a percentage
national_df['hold']=national_df['hold'].apply(lambda x: "{:.2%}".format(x))
#national_df
```

```python
national_df.to_csv('Resources/National_Revenue_Data_CLEAN.csv')
```

```python
# Function to convert a CSV to JSON
# Takes the file paths as arguments
def make_json(csvFilePath, jsonFilePath):
     
    # create a dictionary
    data = {}
     
    # Open a csv reader called DictReader
    with open(csvFilePath, encoding='utf-8') as csvf:
        csvReader = csv.DictReader(csvf)
         
        # Convert each row into a dictionary
        # and add it to data
        for rows in csvReader:
             
            #primary key
            key = rows['']
            data[key] = rows
 
    # Open a json writer, and use the json.dumps()
    # function to dump data
    with open(jsonFilePath, 'w', encoding='utf-8') as jsonf:
        jsonf.write(json.dumps(data, indent=4))
```
