```python
# Dependencies
import requests
import csv
import json
import pandas as pd
from pathlib import Path
```

```python
#read the csvs file into a data frame
team_data=Path('Resources/market_rev_teams_df_FINALSET.csv')
team_df=pd.read_csv(team_data)
team_df
```

```python
#read the csvs file into a data frame
state_data=Path('all_state_market_data_CLEAN.csv')
state_df=pd.read_csv(state_data)
state_df
```

```python
#clean up columns to start with state and eliminate extra column
order=['state','state_abbr', 'handle', 'revenue','hold','taxes']
state_clean_df=state_df[order]
state_clean_df
```

```python
state_aggregated_df=state_clean_df.groupby('state').agg({'handle':'sum', 'revenue':'sum','hold':'mean','taxes':'sum'}).reset_index()
state_aggregated_df
```

```python
#Change handle, revenue and taxes to dollar format
team_df['handle']=team_df['handle'].astype(float).map('${:,.2f}'.format)
team_df['revenue']=team_df['revenue'].astype(float).map('${:,.2f}'.format)
team_df['taxes']=team_df['taxes'].astype(float).map('${:,.2f}'.format)

team_df
```

```python
#Change the hold column to show as a percentage
team_df['hold']=team_df['hold'].apply(lambda x: "{:.2%}".format(x))
team_df
```
