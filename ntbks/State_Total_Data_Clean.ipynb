```python
# Dependencies
import requests
import csv
import json
import pandas as pd
from pathlib import Path
```

```python
#read the csvs file into a data frame
state_total_data= Path('../Resources/all_state_market_data_FINALSET.csv')
state_total_df= pd.read_csv(state_total_data)
#state_total_df
```

```python
#read the csvs file into a data frame
population_data=Path('../Resources/census_population_age_data.csv')
population_df=pd.read_csv(population_data)
#population_df
```

```python
population_df.columns
```

```python
#clean up columns to eliminate extra columns
order=['state', 'population total 18+']
population_df_clean= population_df[order]
#population_df_clean
```

```python
#Merge the Datasets on state
state_total_merge=pd.merge(state_total_df, population_df_clean, on='state', how='inner')
#state_total_merge
```

```python
state_total_merge.dtypes
```

```python
#Changing the population column to a float value
state_total_merge['population total 18+']=state_total_merge['population total 18+'].astype(float)
```

```python
#Add a new column to show the the average reveune per person per state
state_total_merge['Revenue Per Person']= state_total_merge['revenue']/ state_total_merge['population total 18+']
#state_total_merge
```

```python
#Change the date column to datetime format
state_total_merge=state_total_merge.astype({'start_month':'datetime64[D]'})
#state_total_merge
```

```python
#Change handle, revenue and taxes to dollar format
state_total_merge['handle']=state_total_merge['handle'].astype(float).map('${:,.2f}'.format)
state_total_merge['revenue']=state_total_merge['revenue'].astype(float).map('${:,.2f}'.format)
state_total_merge['taxes']=state_total_merge['taxes'].astype(float).map('${:,.2f}'.format)
state_total_merge['population total 18+']=state_total_merge['population total 18+'].astype(float).map('{:,}'.format)
state_total_merge['Revenue Per Person']=state_total_merge['Revenue Per Person'].astype(float).map('${:,.2f}'.format)

#state_total_merge
```

```python
#Change the hold column to show as a percentage
state_total_merge['hold']=state_total_merge['hold'].apply(lambda x: "{:.2%}".format(x))
state_total_merge
```

```python
state_total_merge.to_csv('Resources/State_Total_Data_FINAL.csv')
```

```python
# Function to convert a CSV to JSON
# Takes the file paths as arguments
def make_json(csvFilePath, jsonFilePath):
     
    # create a dictionary
    data = {}
     
    # Open a csv reader called DictReader
    with open(csvFilePath, encoding='utf-8') as csvf:
        csvReader = csv.DictReader(csvf)
         
        # Convert each row into a dictionary
        # and add it to data
        for rows in csvReader:
             
            #primary key
            key = rows['']
            data[key] = rows
 
    # Open a json writer, and use the json.dumps()
    # function to dump data
    with open(jsonFilePath, 'w', encoding='utf-8') as jsonf:
        jsonf.write(json.dumps(data, indent=4))
```
